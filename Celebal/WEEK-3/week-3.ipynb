```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
data = pd.read_csv('annual-enterprise-survey-2021-financial-year-provisional-size-bands.csv')

# Filter the data for 'Total income' variable
total_income_data = data[data['variable'] == 'Total income']

# Summarize the total income by year
total_income_by_year = total_income_data.groupby('year')['value'].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(total_income_by_year['year'], total_income_by_year['value'], marker='o', linestyle='-', color='b')
plt.title('Total Income Over Years')
plt.xlabel('Year')
plt.ylabel('Total Income (in millions of dollars)')
plt.grid(True)
plt.xticks(total_income_by_year['year'])  # Ensures all years are shown on the x-axis
plt.show()
```
